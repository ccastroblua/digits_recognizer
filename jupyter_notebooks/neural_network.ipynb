```python
from PIL import Image
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
import mnist
```

```python
# Train data:

x_train = mnist.train_images()
y_train = mnist.train_labels()
```

```python
# Test data:

x_test = mnist.test_images()
y_test = mnist.test_labels()
```

```python
print('X_train:\n', x_train)
print('X_test:\n', x_test)
```

```python
print('y_train:\n', y_train)
print('y_test:\n', y_test)
```

```python
# How many dimensions does x_train has?

print(x_train.ndim)
```

```python
# 60000 samples of 28x28 pixels:

print(x_train.shape)
```

```python
x_train = x_train.reshape((-1, 28*28))
```

```python
# 60000 samples of 28x28 pixels but now they are in the same dimension:

print(x_train.shape)
```

```python
# Let's do the same with x_test data:

x_test = x_test.reshape((-1, 28*28))
```

```python
x_test.shape
```

```python
# Checking one example of x_train:

print(x_train[0])
```

```python
# Let's rescale it to 0 to 1:

x_train = (x_train/256)
x_test = (x_test/256)
```

```python
# Checking it:

x_train[0]
```

```python
# Initilizing the model:

clf = MLPClassifier(solver='adam', activation='relu', hidden_layer_sizes=(64,64))
```

```python
clf.fit(x_train, y_train)
```

```python
# Testing the model with new predictions

predictions = clf.predict(x_test)
```

```python
# Accuracy:

acc = confusion_matrix(y_test, predictions)
```

```python
print(acc)
```

```python
# The accuracy is the sum of the diagonal numbers divided by 
# the sum of all the number in the matrix:

def accuracy(cm):
    diagonal = cm.trace()
    elements = cm.sum()
    return diagonal/elements
```

```python
print(accuracy(acc))
```
